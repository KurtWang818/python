```python
file = open("../Lecture-03/ques2.txt", "r")

wordLength = {}
count = 0
for word in file.read().split():
    count += 1
    if word not in wordLength:
        length = len(word)
        wordLength[word] = length

file.close()


print(count)

list = wordLength.values()
print("average: " + str(sum(list) / float(len(list))))

median = 0
list = sorted(list)

mid = len(list)//2
if len(list) % 2:
    median = list[mid]
else:
    median = (list[mid] + list[mid + 1]) / 2
print("median: " + str(median))

from collections import Counter

mode = Counter(list)
mode = mode.most_common(1)
print("mode: " + str(mode))
```

```python
import os, json
import random

count = 0
d = {}
path_to_json = '../Lecture-03/jsons'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

for file in json_files:
    with open(os.path.join(path_to_json, file)) as json_file:
        data = json.load(json_file)
        value = data.get("name")
         
        if (value == 'User Not Found'):
            userId = data.get("message")
            if userId is not None:
                userId = userId[13:]
            count += 1
            d[userId] = 0
        else:
            userId = data.get("dxId")
            d[userId] = 1
print(count)
# for (k, v) in d.items():
#     print(k, v)
for i in range(1, 6):
    key = random.choice(d.keys())
    value = d[key]
    print(str(key) + "  " + str(value))
```

```python
import os, json

path_to_json = '../Lecture-03/jsons'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
d = {}

for file in json_files:
    count1 = 0
    count2 = 0
    with open(os.path.join(path_to_json, file)) as json_file:
        data = json.load(json_file)
        segmentUidList = data.get("retargetingSegments")
        
        if segmentUidList is not None:
            userId = data.get("dxId")
            for each in segmentUidList:
                segmentUid = each.get("segmentUid")
                start = int(segmentUid[0])
                if (start == 0):
                    count1 += 1
                    
                elif (start == 1):
                    count2 += 1
    d[userId] = {}
    d[userId]["first"] = count1
    d[userId]["third"] = count2
    
for i in range(1, 6):
    key = random.choice(d.keys())
    value = d[key]
    print(str(key) + str(value))
```
